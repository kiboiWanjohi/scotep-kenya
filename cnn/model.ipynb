```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.models import Model
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Input, Conv2D, BatchNormalization, Activation, Add, Flatten, Dense, Dropout
from tensorflow.keras.optimizers.legacy import Adam, SGD
#from tensorflow.keras.optimizers import Adam, SGD
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import classification_report
import cv2
import os
import numpy as np
```

# Pre-processing

```python
def Normalisation(image):
  normalised_image = (image - image.min())/(image.max() - image.min())
  return(normalised_image)
```

```python
cancer = [ '/content/Cancer_png/' + i for i in os.listdir('/content/Cancer_png') if 'MLO' in i]
benign = [ '/content/Benign_png/' + i for i in os.listdir('/content/Benign_png') if 'MLO' in i]
normal = [ '/content/Normal_png/' + i for i in os.listdir('/content/Normal_png') if 'MLO' in i]
```

```python
print(len(cancer))
print(len(benign))
print(len(normal))
```

```python
x = np.zeros((2000, 360, 200, 3))
y = np.zeros((2000, 1))
for i in range(500):
  x[i] = Normalisation(cv2.resize(cv2.imread(cancer[i]), (200, 360)))
  x[i+500] = Normalisation(cv2.resize(cv2.imread(benign[i]), (200, 360)))
  x[i+1000] = Normalisation(cv2.resize(cv2.imread(normal[i]), (200, 360)))
  x[i+1500] = Normalisation(cv2.resize(cv2.imread(normal[i+500]), (200, 360)))
  if i%100==0:
    print(i)
for i in range(1000):
  y[i] = 1
  y[i + 1000] = 0
y = tf.keras.utils.to_categorical(y)
```

```python
print(x.shape)
print(y.shape)
```

```python
x_, x_test, y_, y_test = train_test_split(x, y, test_size=0.1)
x_train, x_val, y_train, y_val = train_test_split(x_, y_, test_size=0.2)
```

```python
print('x_train shape', len(x_train))
print('x_val shape', len(x_val))
print('x_test shape', len(x_test))

print('y_train shape', len(y_train))
print('y_val shape', len(y_val))
print('y_test shape', len(y_test))
```

# Resnet

```python
Resnet50 = tf.keras.applications.ResNet50(include_top=False, weights=None, input_shape=(360, 200, 3), pooling = 'max')
Resnet50.trainable = True
```

```python
def resnet50_model():
  x = Resnet50.output
  x = Dense(1024, activation='relu')(x)
  x = Dense(512, activation='relu')(x)
  x = Dense(128, activation='relu')(x)
  x = Dense(32, activation='relu')(x)
  x = Dense(8, activation='relu')(x)
  Output = Dense(2, activation='softmax')(x)
  model = Model(inputs=Resnet50.input, outputs=Output)
  return model
model = resnet50_model()
#model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.000001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(x_train, y_train, batch_size=64, epochs=20, validation_data=(x_val, y_val))
```

```python
label_map = {0: 'normal', 1: 'abnormal'}
predict_cnn = model.predict(x_train)
pred_label_model = np.argmax(predict_cnn, axis = 1)

cm = confusion_matrix(np.argmax(y_train, axis = 1), pred_label_model)
cmd = ConfusionMatrixDisplay(cm, display_labels=label_map.values())
print((pred_label_model == np.argmax(y_train, axis = 1)).sum()/len(pred_label_model) * 100)


fig, ax = plt.subplots(figsize=(10, 10))
cmd.plot(ax = ax)
plt.plot('Training Accuracy')
plt.show()
print(classification_report(np.argmax(y_train, axis = 1), pred_label_model, target_names=label_map.values()))
```

```python
label_map = {0: 'normal', 1: 'abnormal'}
predict_cnn = model.predict(x_test)
pred_label_model = np.argmax(predict_cnn, axis = 1)

cm = confusion_matrix(np.argmax(y_test, axis = 1), pred_label_model)
cmd = ConfusionMatrixDisplay(cm, display_labels=label_map.values())
print((pred_label_model == np.argmax(y_test, axis = 1)).sum()/len(pred_label_model) * 100)


fig, ax = plt.subplots(figsize=(10, 10))
cmd.plot(ax = ax)
plt.plot('Testing Accuracy')
plt.show()
print(classification_report(np.argmax(y_test, axis = 1), pred_label_model, target_names=label_map.values()))
```
